```python
# Create a DataFrame of all matches where Brazil participated
brazil_matches = matches[(matches['Home Team Name']=='Brazil')|(matches['Away Team Name']=='Brazil')]
brazil_matches.head()
```

```python
# Create a DataFrame of Brazil's wins, losses and draws in World Cup
brazil_draw = brazil_matches[brazil_matches['Home Team Goals']==brazil_matches['Away Team Goals']]
brazil_draw
```

```python
brazil_win = brazil_matches[((brazil_matches["Home Team Name"] == "Brazil") & (brazil_matches["Home Team Goals"] > brazil_matches["Away Team Goals"])) |
                            ((brazil_matches["Away Team Name"] == "Brazil") & (brazil_matches["Away Team Goals"] > brazil_matches["Home Team Goals"]))]
```

```python
brazil_win
```

```python
brazil_loss = brazil_matches[((brazil_matches["Home Team Name"] == "Brazil") & (brazil_matches["Home Team Goals"] < brazil_matches["Away Team Goals"])) |
                            ((brazil_matches["Away Team Name"] == "Brazil") & (brazil_matches["Away Team Goals"] < brazil_matches["Home Team Goals"]))]
brazil_loss
```

```python
brazil_matches["Outcome"] = ""
brazil_matches.loc[((brazil_matches["Home Team Name"] == "Brazil") & (brazil_matches["Home Team Goals"] > brazil_matches["Away Team Goals"])), "Outcome"] = "Win"
brazil_matches.loc[((brazil_matches["Away Team Name"] == "Brazil") & (brazil_matches["Home Team Goals"] < brazil_matches["Away Team Goals"])), "Outcome"] = "Win"
brazil_matches.loc[((brazil_matches["Home Team Name"] == "Brazil") & (brazil_matches["Home Team Goals"] < brazil_matches["Away Team Goals"])), "Outcome"] = "Loss"
brazil_matches.loc[((brazil_matches["Away Team Name"] == "Brazil") & (brazil_matches["Home Team Goals"] > brazil_matches["Away Team Goals"])), "Outcome"] = "Loss"
brazil_matches.loc[brazil_matches["Outcome"] == "", "Outcome"] = "Draw"
```

```python
# Filter the data to only include matches where Brazil was playing
brazil_score = matches[(matches['Home Team Name'] == 'Brazil') | (matches['Away Team Name'] == 'Brazil')]
```

```python
brazil_score.head()
```

```python
# create score table 
brazil_score['brazil_score_home'] = brazil_score['Home Team Goals'].loc[brazil_score['Home Team Name']=='Brazil']
brazil_score['brazil_score_away'] = brazil_score['Away Team Goals'].loc[brazil_score['Away Team Name']=='Brazil']
brazil_score['brazil_score_home'] = brazil_score['brazil_score_home'].fillna(0)
brazil_score['brazil_score_away'] = brazil_score['brazil_score_away'].fillna(0)
brazil_score['Score'] = brazil_score['brazil_score_home'] + brazil_score['brazil_score_away']
```

```python
brazil_score.head()
```

```python
# create goal conceded table 
brazil_score['brazil_conceded_home'] = brazil_score['Home Team Goals'].loc[brazil_score['Home Team Name']!='Brazil']
brazil_score['brazil_conceded_away'] = brazil_score['Away Team Goals'].loc[brazil_score['Away Team Name']!='Brazil']
brazil_score['brazil_conceded_home'] = brazil_score['brazil_conceded_home'].fillna(0)
brazil_score['brazil_conceded_away'] = brazil_score['brazil_conceded_away'].fillna(0)
brazil_score['Conceded'] = brazil_score['brazil_conceded_home'] + brazil_score['brazil_conceded_away']
```

```python
brazil_score_conceded = brazil_score.groupby('Year')['Score','Conceded'].sum()
brazil_score_conceded
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data =brazil_score_conceded, x=brazil_score_conceded.index, y = 'Score', color='green')
plt.xlabel("Year")
plt.ylabel("Number score of Brazil ")
plt.title("Brazil's Score")
plt.show()
```

```python
# 3. Analyze Brazil's Performance in Different Stages of the World Cup: round 0f 16, quarter final, semi final, final
```

```python
matches['Stage'].unique()
```

```python
# Filter the dataset to only include matches in the rounds of 16, quarter-finals, semi-finals, and final
round_of_16 = brazil_matches[brazil_matches["Stage"] == "Round of 16"]
quarter_finals = brazil_matches[brazil_matches["Stage"] == "Quarter-finals"]
semi_finals = brazil_matches[brazil_matches["Stage"] == "Semi-finals"]
finals = brazil_matches[brazil_matches["Stage"] == "Final"]

# Define a function to calculate the win, loss, and draw percentages
def calculate_percentages(group):
    total_matches = len(group)
    wins = len(group[group["Outcome"] == "Win"])
    losses = len(group[group["Outcome"] == "Loss"])
    draws = total_matches - wins - losses
    win_percentage = (wins / total_matches) * 100
    loss_percentage = (losses / total_matches) * 100
    draw_percentage = (draws / total_matches) * 100
    return win_percentage, loss_percentage, draw_percentage

# Calculate the win, loss, and draw percentages for each round
round_of_16_percentages = calculate_percentages(round_of_16)
quarter_finals_percentages = calculate_percentages(quarter_finals)
semi_finals_percentages = calculate_percentages(semi_finals)
finals_percentages = calculate_percentages(finals)
```

```python
# Create pie charts to visualize the results
labels = ["Win", "Loss", "Draw"]
colors = ["green", "red", "yellow"]

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].pie(round_of_16_percentages, labels=labels, colors=colors, autopct='%1.1f%%')
axs[0, 0].set_title("Round of 16")

axs[0, 1].pie(quarter_finals_percentages, labels=labels, colors=colors, autopct='%1.1f%%')
axs[0, 1].set_title("Quarter-finals")

axs[1, 0].pie(semi_finals_percentages, labels=labels, colors=colors, autopct='%1.1f%%')
axs[1, 0].set_title("Semi-finals")

axs[1, 1].pie(finals_percentages, labels=labels, colors=colors, autopct='%1.1f%%')
axs[1, 1].set_title("Final")

plt.suptitle("Brazil's Performance in Different Stages of the World Cup")
plt.show()
```
